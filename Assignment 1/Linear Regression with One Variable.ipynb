```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading data and assigning variables

data = np.loadtxt("ex1data1.txt", comments="#", delimiter=",")
y = np.array([data[:, 1]]).T
m = np.size(y)
X = np.c_[ np.ones(m), data[:, 0] ]
theta = np.zeros((2,1))
```

```python
#Inital Plot with theta0 = 0 and theta1 = 0

plt.scatter(X[:,1], y[:,0], s=25, c='red')
plt.xlabel('Profit in $10,000s')
plt.ylabel('Population of City in 10,000s')
plt.plot(X[:,1], np.dot(X,theta))
plt.show()
```

```python
# Cost Function

def costFunction(theta):
    hypothesis = np.dot(X,theta)
    squaredError = (hypothesis - y)**2
    sumSquaredError = np.sum(squaredError)
    J = 1/(2*m) * sumSquaredError
    return(J)

print(costFunction(theta))
```

```python
# Gradient Descent

iterations = 1500
alpha = 0.01
jList = []
for i in range(iterations):
    
    beta = np.dot(X,theta)
    delta = (1/m) * np.dot((beta - y).T, X)
    theta = theta - alpha * delta.T
    jList.append(costFunction(theta))
    
print("Gradient Descent Complete with an accuracy of", 100 - costFunction(theta))
```

```python
# Plotting the new hypothesis with converged cost function

plt.scatter(X[:,1], y[:,0], s=25, c='red')
plt.xlabel('Profit in $10,000s')
plt.ylabel('Population of City in 10,000s')
plt.plot(X[:,1], np.dot(X,theta))
plt.show()
```

```python
# Plotting a J vs number of iterations graph to ensure function is always decreasing and has converged

plt.scatter(np.arange(1,1501), jList, s=1, c='red')
plt.xlabel('# of iterations')
plt.ylabel('J (theta0, theta1)')
plt.show()
```

```python
# Using the optimized hypothesis to predict profit with city population of 35,000 and 70000

def usingHypothesis(population):
    prediction = np.dot([1, population/10000], theta)
    return (np.ndarray.item(prediction*10000))

print("For a population of 35000, the ML algorithm predicts a profit of:", usingHypothesis(35000))
print("For a population of 70000, the ML algorithm predicts a profit of:", usingHypothesis(70000))
```
